# Quick wildfire EDA
Just a scratch notebook to poke at the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

damage = pd.read_csv('2018-07-wildfire-trends/data/calfire_damage.csv')
frap = pd.read_csv('2018-07-wildfire-trends/data/calfire_frap.csv')
damage.head()
```

```python
# Quick look at distributions
frap['gis_acres'] = pd.to_numeric(frap['gis_acres'], errors='coerce')
sns.histplot(frap['gis_acres'].dropna(), bins=50)
plt.title('Distribution of GIS acres per fire')
plt.show()
```

```python
# Combine damage and acres to see whether big fire years align with damage
frap['year_'] = pd.to_numeric(frap['year_'], errors='coerce')
frap['report_ac'] = pd.to_numeric(frap['report_ac'], errors='coerce')
frap['gis_acres'] = pd.to_numeric(frap['gis_acres'], errors='coerce')
frap['acres'] = frap['report_ac'].fillna(frap['gis_acres'])
by_year = frap.dropna(subset=['year_', 'acres']).groupby('year_')['acres'].sum().reset_index()
merged = damage.merge(by_year, left_on='year', right_on='year_', how='inner')
sns.scatterplot(data=merged, x='acres', y='structures')
plt.title('Total acres vs. structures damaged (yearly)')
plt.show()
```
